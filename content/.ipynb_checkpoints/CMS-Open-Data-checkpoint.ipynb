# Introduction

The Compact Muon Solenoid (CMS) experiment is one of two large general-purpose particle physics detectors built on the Large Hadron Collider (LHC) at CERN. The goal of CMS experiment is to investigate a wide range of physics, including the search for the Higgs boson, extra dimensions, and particles that could make up dark matter read more.

This open data is released under the Creative Commons CC0 waiver. Neither CMS nor CERN endorse any work, scientific or otherwise, produced using these data.

**The data has been already prepare and contains only event where two (2) muons have been detected.**

Muons are elementary particles from the family of the Leptons, they have greater mass than their electric charge which is −1 e and a spin of 1/2, their mass is actually 200 times heavier than electrons. Because muons can pass through several metres of matter without interacting, they are not stopped by any of the calorimeters on CMS. So chambers to detect muons are placed at the very edge of the detector, where muons are the only particles likely to register a signal [2](https://home.cern/news/news/experiments/watch-engineers-install-muon-detector-cms)

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

dataset_file = '../data/CMS-Open-Data.csv'
dataset = pd.read_csv(dataset_file)
dataset.head()
```

Where 'M' (the last column) is the invariant mass (in GeV) of either two muons or two electrons. the whole semantic is here.

To find out how many instances (rows) and how many attributes (columns) the data contains with the shape property, we can use the print() function and find out the dataset shape:

```python
print(dataset.shape)
```

However, we are only concern in the invariance mass which are the particles that we would like to see so let's visualize the data in terms of this invariant mass:

```python
masses = dataset.M # assigning mass values to a variable 'mass'

# bins from 2.0 to 5.0 in 0.10 intervals
hist = masses.hist(bins=[2.0, 2.10, 2.20, 2.30, 2.40, 2.50, 2.60, 2.70, 2.80, 2.90, 3.0, 3.10, 3.20, 3.30, 3.40, 3.50, 3.60, 3.70, 3.80, 3.90, 4.0, 4.10, 4.20, 4.30, 4.40, 4.50, 4.60, 4.70, 4.80, 4.90, 5.0])
plt.title('The histogram of the invariant masses of two muons \n')
plt.xlabel("Invariant Masses in GeV")
plt.ylabel("Frequency")
plt.show()
```

We could just have defined bins = 4000 or to something else for a quick inspection of the data, however defining it as per above, from 2.0 to 5.0 in step of 0.10, allow us to identify the particle picks.

```python
mean_masses = np.mean(masses)
print(mean_masses) 

variance_masses = np.var(masses)
print(variance_masses)
```

### We conclude then that Two (2) muons were detected and can clearly be seen on the histogram. The big pick at about 3.1 GeV is a decay into two muons, the J/psi meson. The other pick at about 3.7 GeV also decays into two muons and is known to be the Psi meson.
